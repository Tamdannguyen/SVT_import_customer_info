```python
# import libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# import file 
path = 'C:/Users/DELL/Desktop/MẸ ỈN/PROJECT/MIS_ETL_SVTECH/'
df = pd.read_excel(path + 'Số liệu SunViet.xlsx')
df.head()
```

```python
#Get rid of 'NaN' value
df = df.dropna(axis = 1, how = 'all')
df.head()
```

```python
# add 'Giá trị (USD)' and 'Giá trị (VND)' columns
df.insert(16, 'Giá trị (USD)', df['Đơn giá NT khai báo'] * df['Lượng'])
df.insert(17, 'Giá trị (VND)', df['Giá trị (USD)'] * df['Tỷ giá USD'])
df.iloc[:,[16,17]]
```

```python
# transform country name
chars = []
names = []

file = open("C:/Users/DELL/Desktop/MẸ ỈN/PROJECT/MIS_ETL_SVTECH/tenviettat.txt", encoding="utf-8") #Replace the abbreviation with the full name 
name_table = file.read().split("\n")
for name in name_table:
    x = name.split("\t")
    chars.append(x[0])
    names.append(x[1])
    def replace_value(row, chars, names):
        for c, n in zip(chars, names):
            row = row.replace(c, n)
        return row

df.loc[:, 'Tên nuớc xuất xứ'] = df['Tên nuớc xuất xứ'].apply(lambda x: replace_value(x, chars, names))
df.iloc[:,[19]]
```

```python
# clean 'Ngày' column
df.loc[:, 'Ngày hợp đồng'] = pd.to_datetime(df['Ngày hợp đồng'], format='%m/%d/%Y %I:%M:%S %p').dt.date
df.iloc[:,[21]]
```

```python
# transform customs declaration place
chars1 = []
names1 = []

file = open("C:/Users/DELL/Desktop/MẸ ỈN/PROJECT/MIS_ETL_SVTECH/haiquanviettat.txt", encoding="utf-8")
name1_table = file.read().split("\n")
for name in name1_table:
    name = name.strip()
    x = name.split("\t")
    chars1.append(x[0])
    names1.append(x[1])
    def replace_value(row1, chars1, names1):
        for c, n in zip(chars1, names1):
            row1 = row1.replace(c, n)
        return row1

df.loc[:, 'Tên nơi mở tờ khai'] = df['Tên nơi mở tờ khai'].apply(lambda x: replace_value(x, chars1, names1))
df.iloc[:,[2]]
```

```python
# sort into 'mới 100%' and 'đã qua sử dụng'
df.loc[:, 'Phân loại hàng'] = df['Tên hàng'].apply(lambda x: 'Hàng mới 100%' if '100%' in x else 'Hàng đã qua sử dụng')
df.insert(8, 'Phân loại hàng', df.pop('Phân loại hàng'))
df.iloc[:,[8]]
```

```python
# extract product band
df.loc[:, 'Thương hiệu'] = df['Tên hàng'].str.extract(r'hiệu\s+(\w+)')
df.insert(9, 'Thương hiệu', df.pop('Thương hiệu'))
df.iloc[:, [9]]
```

```python
# Add 'Ngày', 'Tháng' and 'Năm' columns
df.loc[:,'Ngày'] = df['Ngày đăng ký'].dt.day
df.loc[:,'Tháng'] = df['Ngày đăng ký'].dt.month
df.loc[:,'Năm'] = df['Ngày đăng ký'].dt.year
df.insert(2, 'Ngày', df.pop('Ngày'))
df.insert(3, 'Tháng', df.pop('Tháng'))
df.insert(4, 'Năm', df.pop('Năm'))
df.iloc[:, [2,3,4]]
```

```python
#check columns' name
print(df.columns)
```

```python
#1. Which year had the highest import revenue, and which month had the highest import revenue in that year?
import_value = df.groupby('Năm')['Giá trị (USD)'].sum()
years = range(2021,2025)
plt.bar(x=years, height=import_value, width = 0.4)
for i, v in enumerate(import_value):
    plt.text(years[i], v + 50, str(v), ha='center', va='bottom')
plt.xticks(years)
plt.xlabel('Năm')
plt.ylabel('Giá trị (USD)')
plt.title('Tổng giá trị nhập khẩu theo năm')
plt.show()
```

```python
# 2. Who is the biggest supplier?
value_per_supplier = df.groupby('Đơn vị đối tác')['Giá trị (USD)'].sum()
top_5_suppliers = value_per_supplier.nlargest(5) # find top 5 biggest suppliers
plt.bar(x=top_5_suppliers.index, height=top_5_suppliers.values, width = 0.4)
plt.xlabel('Đơn vị đối tác')
plt.ylabel('Giá trị (USD)')
plt.title('5 Nhà cung cấp lớn nhất'))
for i, v in enumerate(top_5_suppliers.values):
    plt.text(i, v + 50, str(v), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Which customs clearance location has the highest frequency? 
count_customer_declaration = df.groupby('Tên nơi mở tờ khai').size().nlargest(5)
plt.bar(x=count_customer_declaration.index, height=count_customer_declaration.values, width = 0.4)
plt.xlabel('Tên nơi mở tờ khai')
plt.ylabel('Số lần thông quan')
plt.title('Địa điểm nhập hàng nhiều nhất')
for i, v in enumerate(count_customer_declaration.values):
    plt.text(i, v + 50, str(v), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Exchange rate fluctuation
plt.figure(figsize=(10, 6))
plt.plot(df['Ngày đăng ký'], df['Tỷ giá USD'], label='Tỷ giá')
plt.xlabel('Ngày')
plt.ylabel('Tỷ giá USD')
plt.title('Sự thay đổi của Tỷ giá theo Ngày')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  
plt.show()
```

```python
# 5. Which delivery terms are most commonly used? 
delivery_counts = df.groupby('Điều kiện giao hàng').size()
plt.bar(x=delivery_counts.index, height=delivery_counts.values, width = 0.4)
plt.xlabel('Điều kiện giao hàng')
plt.ylabel('Số lần sử dụng')
plt.title('Điều kiện giao hàng')
for i, v in enumerate(delivery_counts):
    plt.text(i, v + 0.5, str(v), ha='center',va='bottom')
plt.show()
```

```python
# 6. Vehicle types are most commonly used? 
transport_vehicle_counts = df.groupby('Phương tiện vận chuyển').size()
plt.bar(x=transport_vehicle_counts.index, height=transport_vehicle_counts.values, width = 0.4)
plt.xlabel('Phương tiện vận chuyển')
plt.ylabel('Số lần sử dụng')
plt.title('Các loại phương tiện vận chuyển')
plt.xticks(rotation=90)
for i, v in enumerate(transport_vehicle_counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# 7. Which brand's products are most commonly used?
products_counts = df.groupby('Thương hiệu').size().nlargest(10)
plt.bar(x=products_counts.index, height=products_counts.values, width = 0.4)
plt.xlabel('Thương hiệu')
plt.ylabel('Số lần mua')
plt.title('Các thương hiệu hàng hóa')
plt.xticks(rotation=45)
for i, v in enumerate(products_counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# 12. Which country is the most purchased from?
country_sellers_counts = df.groupby('Tên nuớc xuất xứ').size().nlargest(10)
plt.bar(x=country_sellers_counts.index, height=country_sellers_counts.values, width = 0.4)
plt.xlabel('Quốc gia')
plt.ylabel('Số lần mua')
plt.title('Đất nước nhà cung cấp')
plt.xticks(rotation=45)
for i, v in enumerate(country_sellers_counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
